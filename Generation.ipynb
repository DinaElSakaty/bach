```python
!pip install transformers

from transformers import (
    GPT2TokenizerFast,
    DataCollatorForLanguageModeling,
    TextDataset,
    GPT2LMHeadModel,
    TrainingArguments,
    Trainer,
    pipeline)

tokenizer = GPT2TokenizerFast.from_pretrained('aubmindlab/aragpt2-medium')
model = GPT2LMHeadModel.from_pretrained('aubmindlab/aragpt2-medium')
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "train_batch_size": 8, 
    "num_train_epochs": 3,
    "fp16": False,
    "mlm": False,
    }

!pip install simpletransformers
from simpletransformers.language_modeling import LanguageModelingModel

ft_model = LanguageModelingModel('gpt2', 'aubmindlab/aragpt2-medium', args=train_args)
ft_model.tokenizer.add_tokens(["[SCORE 0]", "[SCORE 1]", "[SCORE 2]", "[SCORE 3]"])
ft_model.model.resize_token_embeddings(len(ft_model.tokenizer))
ft_model.train_model("./trainn.txt", eval_file="./testt.txt")
print(ft_model.eval_model("./evall.txt"))
ft_model.tokenizer.save_pretrained("./ar-qa")
ft_model.model.save_pretrained("./ar-qa")

generated=[]
from simpletransformers.language_generation import LanguageGenerationModel
model = LanguageGenerationModel("gpt2", "./ar-qa")

import csv
with open("./sas-ar-cl.csv", "r" , encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        else:
          x=row[3]
          text=row[5]
          if len(x)>0 and len(text)>0:
            line_count += 1
            text=text.split()[0]
          if x== '0.0':
            gen = model.generate('[SCORE 0] ' + text)
            generated.append(gen)
          elif x== '1.0':
            gen = model.generate('[SCORE 1] ' + text)
            generated.append(gen)
          elif x== '2.0':
            gen = model.generate('[SCORE 2] ' + text)
            generated.append(gen)
          elif x== '3.0':
            gen = model.generate('[SCORE 3] ' + text)
            generated.append(gen)

open('./generated.txt', 'w').write("\n".join(generated))

```
